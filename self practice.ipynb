```python
import pandas as pd
# Define a dictionary containing Students data
data = {'Name': ['Jane', 'Princi', 'James', 'Fadi'],
        'Height': [5.1, 6.2, 5.1, 5.2],
        'Qualification': ['Msc', 'MA', 'Msc', 'Msc'],
       'Score 1' : [56,86,77,45],
       'Score 2' : [50,96,60,30]}
df = pd.DataFrame(data)
print(df)
```

```python
df =df.assign(adderss = ['ny','ca','sfo','az'])
df
```

```python
df =df.sort_values(by= 'Score 1', ascending = False)
df
```

```python
df.sort_values(by = ['Score 1', 'Score 2'], ascending=[False, True ])
df
```

```python
#df['Score 1']>70
df =df[df['Score 1']>70]
print(df)
```

```python
df =df[df['Score 1'] <=70]
print(df)
```

```python
df =df[(df['Score 1']>=70) & (df['Score 2']>=80)]
print(df)
```
