```python
def get_starship_crew_and_passengers(ship_name):
    url = f"https://swapi.dev/api/starships/?search={ship_name}"
    response = requests.get(url)
    data = response.json()["results"][0]
    crew = int(data["crew"].replace(',', ''))
    passengers = int(data["passengers"].replace(',', ''))
    return {"crew": crew, "passengers": passengers}

# Example usage:
ship_name = "Star Destroyer"
counts = get_starship_crew_and_passengers(ship_name)
print(f"The {ship_name} has {counts['crew']} crew members and {counts['passengers']} passengers.")
```
